```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
x1 = pd.ExcelFile('Auditing3Way.xlsx')
x1
```

```python
s1 = x1.parse(sheetname='purchase orders')
s2= x1.parse('goods received')
s3 = x1.parse('invoices received')
s1
```

```python
t1 = s1.merge(s2, on='PO Number')
t1
t2 = t1.merge(s3, on= 'PO Number')
t2
```

```python
set(t2['IR Employee'])
```

```python
for row, invoice in t2.iterrows():
    poQu = invoice['PO Quantity']
    poPr = invoice['PO Price']
    grQu = invoice['GR Quantity']
    invQu = invoice['IR Quantity']
    invPr = invoice['IR Price']
    invPaid = round(invoice['IR Value'],2)
    invNum = invoice['IR Number']
    
    supplier = invoice['PO Supplier']
    employee = invoice['IR Employee']
    received = round(grQu * poPr,2)

    isGRQuok = invQu == grQu
    isPOQuok = poQu == grQu
    isPriceok = invPr == poPr
    isPaidok = invPaid == received

    if not isGRQuok: 
        print('GR Qty Anamoly', invNum, supplier, employee, poQu, grQu, invQu, invPr, poPr, received, invPaid, sep="|")
    if not isPOQuok: 
        print('PO Qty Anamoly', invNum, supplier, employee, poQu, grQu, invQu, invPr, poPr, received, invPaid, sep="|")
    if not isPriceok: 
        print('Price Anamoly', invNum, supplier, employee, poQu, grQu, invQu, invPr, poPr, received, invPaid, sep="|")
    if not isPaidok: 
        print('Paid Anamoly', invNum, supplier, employee, poQu, grQu, invQu, invPr, poPr, received, invPaid, sep="|")
```
